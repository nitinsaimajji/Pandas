## Filtering data

```python
new_df=df.loc[ (df['Type 1']=='Grass') & (df['Type 2']=='Fighting')]
new_df.reset_index(drop=True,inplace=True)
new_df
```

## Regex Filtering (filtering based on textual patterns)

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')]
```

```python
import re        ## when re is mentioned with flag we can filter regradless to case sensitive .
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I)]     
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex=True)]
```

## Conditional Changes

```python
df.loc[df['Type 1']=='Fire','Type 1']='Flamer'
df
```

```python
df.loc[df['Total']>500,['Generation','Legendary']]='No Value'
df
```
