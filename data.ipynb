```python
def network_type():
  import pandas as pd
  df=pd.read_excel('phone_data.xlsx')
  df.dropna(axis=1)
  print("enter j")
  j=input()
  A=df.loc[df['network_type']==j]
  df=A.dropna(axis=1)
  df=df[['network','network_type']].drop_duplicates()
  return df
```
