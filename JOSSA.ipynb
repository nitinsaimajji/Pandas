```python
def sort_rank():
  import pandas as pd
  df=pd.read_excel('rank.xlsx')
  df.dropna(axis=1)
  print("enter institute")
  j=input()
  print("Academic program")
  k=input()
  print("quota")
  q=input()
  print("Gender")
  print("seat type")
  s=input()
  print("opening rank")
  r=float(input())
  print("closing rank")
  c=float(input())
  df=df.loc[(df['Institute']==j) & (df['Academic Program Name']==k) & (df['Quota']==q)  & (df['Seat Type']==s)]
  df=df[(df['Opening Rank']>=r)&(df['Closing Rank']<=c)]
  return df
```

```python
df=sort_rank()
df
```
